```python
import requests
import bs4
```

```python
result= requests.get("https://www.navisureka.in")
```

```python
print(result.text)
```

```python
soup=bs4.BeautifulSoup(result.text)
```

```python
print(soup)
```

```python
soup.select("img")
```

```python
a=0
for i in soup.select("img"):
    if i['src'].find("http")==-1:
        continue 
    f=open(f"{a}.jpg","wb")
    print(i['src'])
    f.write(requests.get(i['src']).content)
    f.close()
    a+=1
    print(a)
```
