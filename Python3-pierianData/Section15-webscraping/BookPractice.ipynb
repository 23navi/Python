```python
import requests
import bs4
```

```python
num=1
titleArray=[]
while True:
    
    url=f"http://books.toscrape.com/catalogue/page-{num}.html"
    result=requests.get(url).text
    soup=bs4.BeautifulSoup(result,"lxml")
    

    if soup.select("title")[0].text.find("404")!=-1:
        break
    else:
        a=0
        try:
            #to get the title for the product
            while True:
                title=soup.select(".product_pod h3")[a].text
                titleArray.append(title)
#                 print(title)
                a+=1
        except:
            num+=1
            a=0
#             print(f"PageNumber: {num}, tilteNumber: {a}")
            

print(titleArray)
        
        
```

```python
for i in titleArray:
    print(i)
```
